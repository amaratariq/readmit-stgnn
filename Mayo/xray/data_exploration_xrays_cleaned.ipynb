```python
df_cxr['ADMISSION_DTM'] = pd.to_datetime(df_cxr['ADMISSION_DTM'],format = '%Y-%m-%d %H:%M:%S', errors = 'coerce')
df_cxr['DISCHARGE_DTM'] = pd.to_datetime(df_cxr['DISCHARGE_DTM'],format = '%Y-%m-%d %H:%M:%S', errors = 'coerce')
```

```python
def to_date(x):
    x = str(x)
    return datetime.date(int(x[:4]), int(x[4:6]), int(x[-2:]))
def to_time(x):
    x = str(x)
    if len(x)>4:
        return datetime.time(int(x[:-4]), int(x[-4:-2]), int(x[-2:]))
    elif len(x)>2:
        return datetime.time(int(0), int(x[-4:-2]), int(x[-2:]))
    else:
        return datetime.time(0, 0, int(x[-2:]))
def to_combine(dt, tm):
    return datetime.datetime.combine(dt, tm)


meta['StudyDate_formatted'] = meta['StudyDate'].apply(to_date)
meta['StudyTime_formatted'] = meta['StudyTime'].apply(to_time)
meta['Study_DTM'] = meta.apply(lambda x: to_combine(x.StudyDate_formatted, x.StudyTime_formatted), axis=1)

meta[['Study_DTM', 'StudyDate_formatted', 'StudyDate', 'StudyTime_formatted', 'StudyTime']]
```
